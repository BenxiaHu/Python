# 1. Setup and Sample Data

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Sample DataFrame
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'] * 25,
    'Value1': np.random.randn(100) * 10 + 50,
    'Value2': np.random.randn(100) * 20 + 100,
    'label': np.random.choice(['Low', 'Medium', 'High'], size=100),
    'Date': pd.date_range(start='2023-01-01', periods=100, freq='D')
})

# Display the first few rows of the DataFrame
print(data.head())
```

# 2. Scatter Plot

```python
# Scatter Plot: Value1 vs Value2
sns.set_style('ticks')
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='Value1', y='Value2', hue='label', style='Category', palette='deep')
plt.title('Scatter Plot of Value1 vs Value2')
plt.xlabel('Value1')
plt.ylabel('Value2')
plt.legend(title='label / Value3',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()

# x and y: Define the axes.
# hue: Adds color encoding based on the 'label'.
# style: Differentiates points based on 'Value3'.
# palette: Sets the color palette.

# figsize: Sets the size of the plot.
# title, xlabel, ylabel: Adds titles and labels.
# legend: Customizes the legend.
```

# 3. Line Plot

```python
# Line Plot: Value1 over Date
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(10,6))
sns.lineplot(data=data, x='Date', y='Value1', hue='Category', marker='o')
plt.title('Line Plot of Value1 Over Time by Category')
plt.xlabel('Date')
plt.ylabel('Value1')
plt.legend(title='Category',bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# x: 'Date' to show temporal progression.
# y: 'Value1' to plot.
# hue: Differentiates lines by 'Category'.
# marker: Adds markers to data points.
# rotation: Rotates x-axis labels for better readability.
# tight_layout: Adjusts subplot params for a neat layout.
```

# 4. Bar Plot

```python
# Bar Plot: Mean Value1 by Category
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Category', y='Value1', errorbar='sd', hue='label')
plt.title('Bar Plot of Mean Value1 by Date')
plt.xlabel('Date')
plt.ylabel('Mean Value1')
plt.legend(title='label',bbox_to_anchor=(1, 1))
#plt.xticks()
plt.tight_layout()
plt.show()
```

# 5. Histogram

```python
# Histogram: Distribution of Value1
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
sns.histplot(data=data, x='Value1', bins=30, kde=True, hue='label',palette='deep',multiple='layer')
plt.title('Histogram of Value1')
plt.xlabel('Value1')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

# 6. Box Plot

```python
# Box Plot: Value1 by Category
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
sns.boxplot(data=data, x='Category', y='Value1', hue='label')
plt.title('Box Plot of Value1 by Category')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.legend(title='label',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

# 7. Violin Plot

```python
# Violin Plot: Value2 by Category and Value3
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(10,6))
sns.violinplot(data=data, x='Category', y='Value2', hue='label', split=True, palette='Pastel1')
plt.title('Violin Plot of Value2 by Category and label')
plt.xlabel('Category')
plt.ylabel('Value2')
plt.legend(title='label',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

# 8. Heatmap

```python
# Heatmap: Correlation Matrix
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
corr = data[['Value1', 'Value2']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Value1 and Value2 Correlation')
plt.tight_layout()
plt.show()
```

# 9. Pair Plot

```python
# Pair Plot: Pairwise relationships
sns.set_style('ticks')  # 'white' removes grid lines
sns.pairplot(data, vars=['Value1', 'Value2'], hue='Category', diag_kind='kde', palette='Set1')
plt.suptitle('Pair Plot of Value1 and Value2 by Category', y=1.02)
plt.show()
```

# 10. Joint Plot

```python
# Joint Plot: Value1 vs Value2
sns.set_style('ticks')  # 'white' removes grid lines
sns.jointplot(data=data, x='Value1', y='Value2', kind='scatter', hue='Category', palette='Dark2', height=8)
plt.suptitle('Joint Plot of Value1 vs Value2 by Category', y=1.02)
#plt.legend(bbox_to_anchor=(1, 1))
#plt.tight_layout()
plt.show()
```

# 11. Facet Grid

```python
# Facet Grid: Scatter plots by label
sns.set_style('ticks')  # 'white' removes grid lines
g = sns.FacetGrid(data, col='label', hue='Category', palette='Set2', height=5, aspect=1)
g.map(sns.scatterplot, 'Value1', 'Value2').add_legend()
g.fig.suptitle('Facet Grid of Value1 vs Value2 by label', y=1.05)
plt.show()
```

# 12. Swarm Plot

```python
# Swarm Plot: Value1 by Category
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
sns.swarmplot(data=data, x='Category', y='Value1', hue='label', palette='Set3', dodge=True)
plt.title('Swarm Plot of Value1 by Category and label')
plt.xlabel('Category')
plt.ylabel('Value1')
plt.legend(title='label',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

# 13. Strip Plot

```python
# Strip Plot: Value2 by Category
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
sns.stripplot(data=data, x='Category', y='Value2', hue='label', palette='Set1', jitter=True, dodge=True)
plt.title('Strip Plot of Value2 by Category and label')
plt.xlabel('Category')
plt.ylabel('Value2')
plt.legend(title='label',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

# 14. Count Plot

```python
# Count Plot: Frequency of Categories
sns.set_style('ticks')  # 'white' removes grid lines
plt.figure(figsize=(8,6))
sns.countplot(data=data, x='Category', hue='label', palette='Accent')
plt.title('Count Plot of Categories by label')
plt.xlabel('Category')
plt.ylabel('Count')
plt.legend(title='label',bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
